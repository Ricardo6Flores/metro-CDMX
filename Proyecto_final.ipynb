## Esta función hace que podamos localizar en un mapa cartesiano las coordenadas de las estaciones de metro

```python
def geo_to_cartesian(lat, lon):
    lat_min = 19.0   # Latitud mínima (en grados decimales)
    lat_max = 20.0   # Latitud máxima (en grados decimales)
    lon_min = -100.0 # Longitud mínima (en grados decimales)
    lon_max = -98.0  # Longitud máxima (en grados decimales)

    # Normalizar las coordenadas para el gráfico
    x = (lon - lon_min) / (lon_max - lon_min)
    y = (lat - lat_min) / (lat_max - lat_min)
    
    return (x, y)
```

# Construimos el grafo estructurado de manera distinta

```python
def convertir_a_grafo(pesos):
    grafo = {}
    
    # Iterar sobre cada arista (origen, destino, peso)
    for origen, destino, _ in pesos:
        if origen not in grafo:
            grafo[origen] = []
        if destino not in grafo:
            grafo[destino] = []
        
        # Añadir ambos nodos a sus listas de adyacencia
        grafo[origen].append(destino)
        grafo[destino].append(origen)
    
    return grafo

# Llamar a la función
grafo = convertir_a_grafo(pesos_B + pesos_L3 + pesos_L2 + pesos_L1 + pesos_L8)
```

# El algoritmo genetico

```python
inicio = "Villa de Aragon"
fin = "Popotla"
tam_poblacion = 15
num_generaciones = 100

# Función para generar una ruta aleatoria
def generar_ruta(grafo, inicio, fin):
    ruta = [inicio]
    nodo_actual = inicio
    max_pasos = 100  # Límite para evitar ciclos infinitos
    pasos = 0

    while nodo_actual != fin and pasos < max_pasos:
        siguientes_nodos = [nodo for nodo in grafo[nodo_actual] if nodo not in ruta]
        if not siguientes_nodos:  # Si no hay nodos disponibles, reinicia
            return generar_ruta(grafo, inicio, fin)
        siguiente_nodo = random.choice(siguientes_nodos)
        ruta.append(siguiente_nodo)
        nodo_actual = siguiente_nodo
        pasos += 1

    if nodo_actual != fin:  # Si no se llegó al fin, reinicia
        return generar_ruta(grafo, inicio, fin)
    return ruta

# Función de fitness (queremos minimizar la longitud)
def fitness(ruta):
    return 1 / len(ruta)  # Cuanto más corta, mejor

# Cruza (crossover) entre dos rutas
def cruzar(ruta1, ruta2):
    punto = len(ruta1) // 2
    nueva_ruta = ruta1[:punto]
    for nodo in ruta2:
        if nodo not in nueva_ruta:
            nueva_ruta.append(nodo)
        if nodo == fin:
            break
    return nueva_ruta

# Mutación: cambia un nodo por otro aleatorio conectado
def mutar(ruta, grafo):
    if len(ruta) > 2:
        idx = random.randint(1, len(ruta) - 2)  # No cambiar inicio ni fin
        nodo_actual = ruta[idx]
        vecinos = grafo[ruta[idx - 1]]
        if vecinos:
            ruta[idx] = random.choice(vecinos)
    return ruta

# Inicializar población
poblacion = [generar_ruta(grafo, inicio, fin) for _ in range(tam_poblacion)]
#print(poblacion)
# Algoritmo Genético
for _ in range(num_generaciones):
    # Evaluación
    poblacion = sorted(poblacion, key=fitness, reverse=True)
    
    # Selección de los mejores
    nueva_poblacion = poblacion[:tam_poblacion//2]

    # Cruza y mutación
    while len(nueva_poblacion) < tam_poblacion:
        p1, p2 = random.sample(nueva_poblacion, 2)
        hijo = cruzar(p1, p2)
        if random.random() < 0.1:  # Probabilidad de mutación
            hijo = mutar(hijo, grafo)
        nueva_poblacion.append(hijo)

    poblacion = nueva_poblacion
    #print(_)
# Mejor solución encontrada
mejor_ruta = min(poblacion, key=len)
print("Mejor ruta encontrada:", mejor_ruta)
```
